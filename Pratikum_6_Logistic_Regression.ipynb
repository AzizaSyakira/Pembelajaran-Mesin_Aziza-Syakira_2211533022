```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_score, recall_score, f1_score
```

Bagian ini mengimpor pustaka-pustaka yang diperlukan untuk analisis data, termasuk pandas untuk manipulasi data, numpy untuk operasi numerik, dan beberapa modul dari scikit-learn untuk membagi data, membangun model regresi logistik, dan mengevaluasi kinerja model.

```python
# Membaca data dari file CSV dan menampilkannya
df = pd.read_csv('cancer patient data sets.csv')
df.head()
```

Bagian ini membaca dataset dari file CSV menggunakan pandas dan menampilkan beberapa baris pertama untuk memberikan gambaran awal tentang data yang dimuat.

```python
df.columns
```

Bagian ini menampilkan informasi umum tentang dataset,yaitu nama-nama kolom dalam dataset.

```python
df=df[['Age', 'Gender', 'Air Pollution', 'Alcohol use',
       'Dust Allergy', 'OccuPational Hazards', 'Genetic Risk',
       'chronic Lung Disease', 'Balanced Diet', 'Obesity', 'Smoking',
       'Passive Smoker', 'Chest Pain', 'Coughing of Blood', 'Fatigue',
       'Weight Loss', 'Shortness of Breath', 'Wheezing',
       'Swallowing Difficulty', 'Clubbing of Finger Nails', 'Frequent Cold',
       'Dry Cough', 'Snoring', 'Level']]
```

Bagian ini memilih subset kolom yang relevan dari dataset untuk analisis lebih lanjut dan menampilkan beberapa baris pertama dari data yang sudah dipilih.

```python
df['Level'].value_counts()
df = df.replace({'Level': {'Low': 1, 'Medium': 2, 'High': 3}})
df.head()
```

Bagian ini mengubah nilai kategorikal pada kolom 'Level' menjadi nilai numerik agar bisa digunakan dalam analisis dan pemodelan.

```python
# Memilih fitur (X) dan target (y) dari dataset
X = df.drop('Level', axis=1)
y = df['Level']
```

Bagian ini memisahkan fitur (X) yang akan digunakan untuk prediksi dan target (y) yang akan diprediksi dari dataset.

```python
# Membagi dataset menjadi set pelatihan dan pengujian
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

Bagian ini membagi dataset menjadi set pelatihan dan pengujian dengan rasio 70:30 menggunakan fungsi train_test_split.

```python
# Membuat model regresi logistik dan melatihnya dengan data pelatihan
Classifier = LogisticRegression(solver='liblinear')
Classifier.fit(X_train, y_train)
```

Bagian ini membuat model regresi logistik dan melatihnya menggunakan data pelatihan.

```python
# Membuat prediksi menggunakan data pengujian
y_test_hat = Classifier.predict(X_test)

# Membuat DataFrame untuk membandingkan nilai aktual dan prediksi
Results = pd.DataFrame({'Actual': y_test, 'Predictions': y_test_hat})
Results.head(10)
```

Bagian ini membuat prediksi menggunakan data pengujian, kemudian menampilkan hasil prediksi bersama dengan nilai aktual dalam sebuah DataFrame.

```python
# Menghitung akurasi model pada data pelatihan
print(accuracy_score(y_test, y_test_hat))
y_train_hat = Classifier.predict(X_train)
print(accuracy_score(y_train, y_train_hat))
```

Bagian ini menghitung akurasi model pada data pengujian dan data pelatihan untuk menilai kinerja model.

```python
# Memeriksa probabilitas prediksi model
y_test_hat_proba = Classifier.predict_proba(X_test)
print(y_test_hat_proba.shape)
y_test_hat_proba[0:5, :]
```

Bagian ini memeriksa probabilitas prediksi dari model dan menampilkan beberapa baris pertama dari DataFrame yang membandingkan nilai aktual, prediksi, dan probabilitas prediksi.

```python
# Membuat DataFrame untuk membandingkan nilai aktual, prediksi, dan probabilitas prediksi
pls = y_test_hat_proba[:, 1]
Results = pd.DataFrame({'Actual': y_test, 'Predictions': y_test_hat, 'Prob(Class = 3)': pls})
Results.head(5)
```

Bagian ini membuat DataFrame 'result' untuk membandingkan nilai aktual, prediksi, dan probabilitas kelas 3.

```python
# Fungsi untuk menampilkan metrik evaluasi model
def perform(y_pred):
    print("Precision : ", precision_score(y_test, y_test_hat, average='micro'))
    print("Recall : ", recall_score(y_test, y_test_hat, average='micro'))
    print("Accuracy : ", accuracy_score(y_test, y_test_hat))
    print("F1 Score : ", f1_score(y_test, y_test_hat, average='micro'))
    cm = confusion_matrix(y_test, y_pred)
    print("\n", cm)
    print("\n")
    print("**" * 27 + "\n" + " " * 16 + "Classification Report\n" + "**" * 27)
    print(classification_report(y_test, y_test_hat))
    print("**" * 27 + "\n")

perform(y_test_hat)
```

Bagian ini mendefinisikan fungsi perform yang menghitung dan menampilkan metrik evaluasi model seperti precision, recall, accuracy, F1 score, serta classification report untuk memberikan rincian lebih lanjut tentang kinerja model pada setiap kelas.
